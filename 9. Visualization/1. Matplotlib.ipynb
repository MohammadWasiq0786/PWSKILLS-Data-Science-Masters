```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x= np.linspace(0, 10, 200)
x
```

```python
y= np.sin(x)
y
```

```python
plt.plot(x, y)
```

```python
plt.plot(x,y)
plt.xlabel("This is my x Data")
plt.ylabel("This is my y Data")
plt.title("This is my Title ")
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x, y)
```

```python
x
```

```python
y
```

```python
colours= np.random.rand(50)
sizes= 1000* np.random.rand(50)

plt.scatter(x, y, c=colours, s= sizes, alpha= 0.5)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Random Rata")
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
colours = np.random.rand(50)
sizes = 1000* np.random.rand(50)

plt.scatter(x, y, c= '#A64D13', s= sizes, alpha=0.5, marker='v')
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Random Data")
```
