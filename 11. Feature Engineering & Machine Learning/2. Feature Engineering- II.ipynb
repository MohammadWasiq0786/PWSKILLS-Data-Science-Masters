## 1.1 Feature Scaling
Feature scaling is a technique used to standardize the range of input variables or features in a dataset. It involves transforming the data so that it is on the same scale and has a similar range. Feature scaling is important because it can improve the performance of machine learning algorithms that use distance-based metrics or optimization algorithms that are sensitive to feature scales.

There are several methods for scaling features, including:

### 1.1.1 Standardization
This method scales the data to have zero mean and unit variance. Each feature is transformed to have a mean of zero and a standard deviation of one.

$$Standardization= \frac{(x - \bar{x})}{std(x)}$$

where,
<br> $x$ is the original feature value
<br> $\bar{x}$ is the mean of the feature values
<br> $std(x)$ is the standard deviation of the feature values.


### 1.1.2 Min-Max Scaling
This method scales the data to a fixed range, usually between 0 and 1. Each feature is transformed to have a minimum value of 0 and a maximum value of 1.

$$Min-Max Scaling= \frac{(x - min(x))} {(max(x) - min(x))}$$

where,
<br> $x$ is the original feature value
<br> $min(x)$ is the minimum value of the feature values <br> $max(x)$ is the maximum value of the feature values.

### 1.1.3 Max Abs Scaling
This method scales the data so that the maximum absolute value of each feature is 1.

$$Max Abs Scaling= \frac{x} {max(abs(x))}$$

where,
<br> $x$ is the original feature value 
<br> $max(abs(x))$ is the maximum absolute value of the feature values.

### 1.1.4 Robust Scaling
This method scales the data to be robust to outliers. It uses the interquartile range to scale the data so that it has a median of zero and a range that is proportional to the interquartile range.

$$Robust Scaling= \frac{(x - median(x))} {iqr(x)}$$

where,
<br> $x$ is the original feature value
<br> $median(x)$ is the median of the feature values
<br> $iqr(x)$ is the interquartile range of the feature values.


The choice of scaling method depends on the specific characteristics of the data and the requirements of the machine learning algorithm being used. In general, it is a good practice to scale the data before using it for machine learning to ensure that the algorithm performs optimally.
